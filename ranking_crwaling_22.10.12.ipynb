```python
# 날짜와 제목, 평점을 저장한 빈 리스트를 생성

import time

movie_date =[]
movie_name = []
movie_point = []
```

```python
# 한 페이지에서 만들어지는 정보를 extend 시키며 날짜를 돌려가자

for today in date:
    html ="https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date={date}"
    response = urlopen(html.format(date = today.strftime('%Y%m%d')))
    soup = BeautifulSoup(response, "html.parser")
    
    movie_date.extend([today]*len(soup.find_all('td', 'point')))
    movie_name.extend([each.a.string for each in soup.find_all('div','tit5')])
    movie_point.extend([each.string for each in soup.find_all('td','point')])
    
    print(str(today))
    time.sleep(1)
    
# time.sleep(1)을 주는 이유는 한개씩 업로드 하는 시간을 1초 간격으로 주는 것이다.
```

```python
len(movie_date),len(movie_name), len(movie_point)

# 세 리스트의 크기가 같음으로 문제가 없다
```

```python
# pandas의 DataFrame으로 만들자

movie = pd.DataFrame({'date': movie_date, 'name':movie_name, 'point':movie_point})
movie.head()
```

```python
# 각 컬럼별 타입 확인 및 null 값 유무 확인을 위해 info 확인
movie.info()
```

> 위 info를 확인해보면 평점 part가 object 즉 글자형으로 되어있다. \
> 추후에 정렬기능을 사용할 예정임으로 이를 숫자형 float형으로 변환하자

```python
# 평점을 float형으로 변환한 후 다시 info 확인
movie['point'] = movie['point'].astype(float)
movie.info()
```

```python
movie.to_csv("./data/naver_movie221012.csv", sep=",", encoding="utf-8")
```

```python
# 위에서 저장한 csv 파일을 읽어보자!!

import numpy as np
import pandas as pd

movie = pd.read_csv("./data/naver_movie221012.csv", index_col = 0)
movie.head(10)
```

```python
# 최근 1개월간 가장 인기 있는 영화를 찾기 위해 날짜별 평점을 더해보겠다!! 
# pivot 테이블 활용

movie_unique = pd.pivot_table(movie, index =['name'], aggfunc=np.mean)
movie_unique.head()
```

```python
# 최근 1개월간 평점이 가장 높은 영화 top 10

movie_best = movie_unique.sort_values(by='point', ascending=False)
movie_best.head(10)
```
