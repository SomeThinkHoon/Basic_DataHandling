```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service

chrome_options = webdriver.ChromeOptions()
driver= webdriver.Chrome(service=Service('./driver/chromedriver'), options=chrome_options)
driver.get('https://www.selenium.dev/')
```

```python
# url 가져오기

driver.current_url
```

```python
# 맨 밑으로 스크롤 하기

driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")

# scrollTo (좌우, 위아래) 이동
```

```python
# 맨 위로 스크롤 하기 

driver.execute_script("window.scrollTo(0,0)")
```

```python
driver.save_screenshot('./data/selenium_capture_1.png')
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys


xpath = '''//*[@id="docsearch"]/button'''

search_btn = driver.find_element(By.XPATH, xpath).click()
```

---
x path는 편하지만, 홉페이지가 업데이트 될 경우 xpath가 변경된다.
---

```python
css_selector = '''#docsearch-input'''
search_docs = driver.find_element(By.CSS_SELECTOR, css_selector)
search_docs.send_keys('CSS Selector')
```

```python
search_docs.send_keys(Keys.ENTER)
```

```python
about = driver.find_element(By.CLASS_NAME, 'nav-item.dropdown').click()
```

```python
from selenium.webdriver import ActionChains

about = driver.find_element(By.XPATH, '''//*[@id="navbarDropdown"]''')
language = driver.find_element(By.XPATH, '''//*[@id="navbarDropdown"]''')

ActionChains(driver).click(about).click(language).perform()
```
