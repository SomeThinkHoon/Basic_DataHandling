```python
# import

from bs4 import BeautifulSoup
from urllib.request import urlopen
import urllib
from urllib.request import Request
```

# 네이버 영화 순위 사이트의 링크 soup에 입력 후 확인

```python
url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=pnt&date=20210808"
page = urlopen(url)

soup = BeautifulSoup(page, "html.parser")

print(soup.prettify())
```

# <font color = red>퀴즈 1번!!</font>

### 구분 가능한 가장 낮은 depth가 아래와 같아 보여 '제목'과 '평점'을 find 시킴

- 이후에 a태그까지 더 크기를 줄이지만, 그 전에 해당 부분에서도 가능하도록 고민을 해보아야겠다.

```python
title = soup.find_all("div", {"class" : "tit5"})
```

```python
point = soup.find_all("td", {"class" : "point"})
```

> 제목과 평점이 잘 보이는지 확인을 위해 이전에 사용했던 반복문 사용

```python
n = 1
for each in title:
    print("[" + str(n) + "] ================================")
    print(each)
    n = n + 1
```

```python
n = 1
for each in point:
    print("[" + str(n) + "] ================================")
    print(each)
    n = n + 1
```

```python
osh = []
for rank in range(0,50,1):
    a = title[rank].find('a').string  # 최하위까지 가도록
    b = point[rank].text
    movie_list = [a, b]
    osh.extend(movie_list)
print(osh)


# text 보다는 get_text 혹은 string 사용!! 용도에 맞게 사용
```

```python
# title과 depth가 같으므로 변수명만 바꿔 그대로 다시 불러왔다.

link = soup.find_all("div", {"class" : "tit5"})

print(link)
```
